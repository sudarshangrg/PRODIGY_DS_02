### Importing necessary python libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the dataset file

```python
data = pd.read_csv('Titanic.csv')
```

```python
data.drop(columns='Cabin').isna().mean()
```

```python
data = data.drop(columns='Cabin').dropna(subset=[Embarked])
```

```python
data['Age'] = data['Age'].round()

data['Pclass'] = data['Pclass'].map({1:'Upper', 2:'Middle', 3:'Lower'})

data['Embarked'] = data['Embarked'].map({1:'Upper', 2:'Middle', 3:'Lower'})

data['Survived'] = data['Survived'].map({0:'Survived', 1:'Not Survived'})

data['Sex'] = data['Sex'].str.title()

categorical_columns = ['Sex', 'Parch', 'SibSp', 'Pclass', 'Embarked', 'Survived']
data[categorical_columns] = data[categorical_columns].astype('category')
```

```python
cols = [
    'Name', 'Sex', 'Age', 'Parch', 'SibSp',
    'Ticket', 'Pclass', 'Embarked', 'Fare',
    'Survived'
]

data = data[cols]
```

```python
data.info
```

### Exploring Data

```python
sns.countplot(data=data, x='Survived')

plt.title('Suvived v/s Not Survived')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))

sns.boxplot(data=data, x='Age', y='Survived', ax=ax1, palette='muted')
ax1.set_xlabel('Age', fontsize=14)
ax1.set_ylabel('Survived', fontsize=14)
ax1.set_title('Distribution of Age')

sns.boxplot(data=data, x='Fare', y='Survived', ax=ax2, palette='muted')
ax2.set_xlabel('Fare', fontsize=14)
ax2.set_ylabel('Survived', fontsize=14)
ax2.set_title('Distribution of Fare')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data=data, hue='Survived', aspect=2)
```

```python
cat_cols = ['Sex', 'Pclass', 'Parch', 'SibSp']
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

for feature, ax in zip(cat_cols, axes.flatten()):
    sns.countplot(data=data, x=feature, hue='Survived', ax=ax)
    ax.set_title(f'Count of {feature}')
    ax.set_xlabel('')
    ax.set_ylabel('')

plt.tight_layout()
plt.show()
```
